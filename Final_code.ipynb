```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
import os
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, classification_report, 
    confusion_matrix, f1_score, precision_recall_curve,
    average_precision_score
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import (
    RandomForestClassifier,
    VotingClassifier
)
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.utils.class_weight import compute_class_weight
import warnings

warnings.filterwarnings('ignore')

np.random.seed(42)
def load_and_preprocess(train_path, test_path):
    train_data = pd.read_csv(train_path)
    test_data = pd.read_csv(test_path)
    
    train_data['service'] = train_data['service'].replace('-', 'unknown')
    test_data['service'] = test_data['service'].replace('-', 'unknown')
    
    cols_to_drop = ['id', 'is_sm_ips_ports', 'ct_flw_http_mthd', 'is_ftp_login']
    train_data.drop(columns=cols_to_drop, inplace=True, errors='ignore')
    test_data.drop(columns=cols_to_drop, inplace=True, errors='ignore')
    
    return train_data, test_data


def feature_engineering(train_data, test_data):
    cat_cols = ['proto', 'service', 'state']
    train_data = pd.get_dummies(train_data, columns=cat_cols)
    test_data = pd.get_dummies(test_data, columns=cat_cols)
    
    for col in train_data.columns.difference(test_data.columns):
        test_data[col] = 0
    for col in test_data.columns.difference(train_data.columns):
        train_data[col] = 0
        
    test_data = test_data[train_data.columns]
    
    num_cols = train_data.select_dtypes(include=['int64', 'float64']).columns
    num_cols = num_cols.drop(['label'], errors='ignore')
    
    scaler = MinMaxScaler()
    train_data[num_cols] = scaler.fit_transform(train_data[num_cols])
    test_data[num_cols] = scaler.transform(test_data[num_cols])
    
    for df in [train_data, test_data]:
        df['packet_ratio'] = df['spkts'] / (df['dpkts'] + 1e-6)
        df['byte_ratio'] = df['sbytes'] / (df['dbytes'] + 1e-6)
        df['duration_per_packet'] = df['dur'] / (df['spkts'] + df['dpkts'] + 1e-6)
        df['response_ratio'] = df['response_body_len'] / (df['sbytes'] + 1e-6)
    
    return train_data, test_data


def encode_labels(y_train, y_test):
    le = LabelEncoder()
    y_train_encoded = le.fit_transform(y_train)
    
    unseen_mask = ~y_test.isin(le.classes_)
    if unseen_mask.any():
        print(f"Note: {unseen_mask.sum()} samples with unseen labels mapped to 'Other'")
        y_test = y_test.where(~unseen_mask, 'Other')
        le.classes_ = np.append(le.classes_, 'Other')
    
    y_test_encoded = le.transform(y_test)
    return y_train_encoded, y_test_encoded, le


def train_and_evaluate(X_train, X_test, y_train, y_test, problem_type='binary'):
    classes = np.unique(y_train)
    weights = compute_class_weight('balanced', classes=classes, y=y_train)
    class_weights = dict(zip(classes, weights))
    
    models = {
        "Balanced_Logistic_Regression": LogisticRegression(
            max_iter=1000, 
            class_weight=class_weights,
            n_jobs=-1
        ),
        "XGBoost": XGBClassifier(
            scale_pos_weight=len(y_train[y_train==0])/len(y_train[y_train==1]),
            use_label_encoder=False,
            eval_metric='logloss',
            n_jobs=-1
        ),
        "LightGBM": LGBMClassifier(
            class_weight='balanced',
            objective='binary' if problem_type == 'binary' else 'multiclass',
            n_jobs=-1
        ),
        "Balanced_Random_Forest": RandomForestClassifier(
            class_weight=class_weights,
            n_estimators=200,
            n_jobs=-1
        ),
        "Voting_Classifier": VotingClassifier(
            estimators=[
                ('xgb', XGBClassifier(use_label_encoder=False, eval_metric='logloss', n_jobs=-1)),
                ('lgbm', LGBMClassifier(class_weight='balanced', n_jobs=-1)),
                ('rf', RandomForestClassifier(class_weight=class_weights, n_jobs=-1))
            ],
            voting='soft',
            n_jobs=-1
        )
    }
    
    if problem_type == 'multiclass':
        models = {
            name: Pipeline([
                ('smote', SMOTE(random_state=42)),
                ('model', model)
            ]) for name, model in models.items()
        }
    
    results = {}
    for name, model in models.items():
        print(f"\nTraining {name}...")
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:, 1] if problem_type == 'binary' else None
        
        accuracy = accuracy_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred, average='weighted')
        report = classification_report(y_test, y_pred)
        
        results[name] = {
            'model': model,
            'accuracy': accuracy,
            'f1_score': f1,
            'report': report,
            'y_proba': y_proba
        }
        
        print(f"\n{name} Results:")
        print(f"Accuracy: {accuracy:.4f}")
        print(f"F1 Score: {f1:.4f}")
        print(report)
        
        plot_confusion_matrix(y_test, y_pred, name)
        
        if problem_type == 'binary':
            plot_precision_recall(y_test, y_proba, name)
    
    return results


def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

def plot_precision_recall(y_true, y_proba, model_name):
    precision, recall, _ = precision_recall_curve(y_true, y_proba)
    ap = average_precision_score(y_true, y_proba)
    
    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, label=f'{model_name} (AP = {ap:.2f})')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.legend()
    plt.show()


def main():
    train_path = "UNSW_NB15_training-set.csv"
    test_path = "UNSW_NB15_testing-set.csv"
    
    print("Loading and preprocessing data...")
    train_data, test_data = load_and_preprocess(train_path, test_path)
    
    print("\nPerforming enhanced feature engineering...")
    train_data, test_data = feature_engineering(train_data, test_data)
    
    X_train = train_data.drop(columns=['label', 'attack_cat'])
    y_train_bin = train_data['label']
    y_train_multi = train_data['attack_cat']
    
    X_test = test_data.drop(columns=['label', 'attack_cat'])
    y_test_bin = test_data['label']
    y_test_multi = test_data['attack_cat']
    
    print("\n=== Enhanced Binary Classification ===")
    binary_results = train_and_evaluate(
        X_train, X_test, y_train_bin, y_test_bin, 'binary'
    )
    
    print("\n=== Enhanced Multi-class Classification ===")
    y_train_multi_encoded, y_test_multi_encoded, le = encode_labels(
        y_train_multi, y_test_multi
    )
    
    le.classes_ = np.array(['Normal', 'Generic', 'Exploits', 'DoS', 'Reconnaissance', 
                          'Shellcode', 'Backdoor', 'Analysis', 'Other'])
    
    X_test = X_test.loc[y_test_multi.index]
    
    multiclass_results = train_and_evaluate(
        X_train, X_test, 
        y_train_multi_encoded, y_test_multi_encoded, 
        'multiclass'
    )
    
    print("\nSaving enhanced models...")
    os.makedirs('no_svm_models', exist_ok=True)
    
    for name, result in binary_results.items():
        with open(f'no_svm_models/{name.lower()}_binary.pkl', 'wb') as f:
            pickle.dump(result['model'], f)
    
    for name, result in multiclass_results.items():
        with open(f'no_svm_models/{name.lower()}_multiclass.pkl', 'wb') as f:
            pickle.dump((result['model'], le), f)
    
    print("\nAll enhanced models saved successfully!")

if __name__ == "__main__":
    main()
```
